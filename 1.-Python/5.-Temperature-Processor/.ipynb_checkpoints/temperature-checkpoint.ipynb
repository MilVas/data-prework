As changing the cooling system is expensive and you are an excellent data analyst, you can't make a decision without basis.

## Tasks
The temperatures measured throughout the 24 hours of a day are:

```python
temperatures_C = [33, 66, 65, 0, 59, 60, 62, 64, 70, 76, 80, 81, 80, 83, 90, 79, 61, 53, 50, 49, 53, 48, 45, 39]
```

The first element of the list is the temperature at 12am, the second element is the temperature at 1am, and so on. 

The company has decided that if one of the following events occurs, then the cooling system needs to be replaced for a new one to avoid damaging the processor.
* More than 4 temperatures are greater than or equal to 70ÂºC.
* Any temperature is above 80ÂºC.
* The average temperature exceeds 65ÂºC.

Follow the steps so that you can make the decision.

#### 1. Find the minimum temperature of the day and store it in a variable.

```python
temperatures_C = [33, 66, 65, 0, 59, 60, 62, 64, 70, 76, 80, 81, 80, 83, 90, 79, 61, 53, 50, 49, 53, 48, 45, 39]
min_temp = min(temperatures_C)
print(min_temp)
```

#### 2. Find the maximum temperature of the day and store it in a variable.

```python
max_temp = max(temperatures_C)
print(max_temp)
```

#### 3. Create a list with the temperatures that are greater than or equal to 70ÂºC. Store it in a variable.

```python
high_temp = []

for i in temperatures_C:
    if i >= 70:
        high_temp.append(i)
print(high_temp)
```

#### 4. Find the average temperature of the day and store it in a variable.

```python
avg_temp = sum(temperatures_C)/len(temperatures_C)
print(avg_temp)
```

#### 5. Imagine that there was a sensor failure at 3am and the data for that specific hour was not recorded. How would you estimate the missing value? Replace the current value of the list at 3am for an estimation. 

```python
temperatures_C = [33, 66, 65, 0, 59, 60, 62, 64, 70, 76, 80, 81, 80, 83, 90, 79, 61, 53, 50, 49, 53, 48, 45, 39]
print(temperatures_C)
temperatures = []
for i in temperatures_C:
    temperatures.append(i)
temperatures.sort()
print(temperatures)
len(temperatures)
median_temp = (temperatures[10] + temperatures[11])/2
print(median_temp)

temperatures_C[3] = median_temp
print(temperatures_C)


```

#### 6. Bonus: the maintenance staff is from the United States and does not understand the international metric system. Help them by converting the temperatures from Celsius to Fahrenheit.
To know more about temperature conversion check this [link](https://en.wikipedia.org/wiki/Conversion_of_units_of_temperature).

**Formula**: 

$F = 1.8 * C + 32$

```python
temperatures_F = []
for i in temperatures_C:
    temperatures_F.append(i*1.8 + 32)
print(temperatures_F)
print(temperatures_C)
```

#### 4. Find the average value of the temperature lists (ÂºC and ÂºF). What is the relation between both average values?

```python
# ğ¹=1.8âˆ—ğ¶+32

print(temperatures_C)
print(temperatures_F)
    
avg_temp_C = avg_temp
avg_temp_F = sum(temperatures_F)/len(temperatures_F)

print("The average temperature is" , round(avg_temp_C,2), "C.")
print("The average temperature is" , round(avg_temp_F,2), "F.")

if round(avg_temp_F,2) < round(1.8 * avg_temp_C + 32, 2) :
    print("The average temperature in Celsius is higher than the average temperature in Fahrenheit, when the Celsius values are converted to Fahrenheit.")
elif round(avg_temp_F,2) == round(1.8 * avg_temp_C + 32, 2):
    print("They are exactly the same, when the Celsius values are converted to Fahrenheit.")
elif round(avg_temp_F,2) > round(1.8 * avg_temp_C + 32, 2) :
    print("The average temperature in Fahrenheit is higher than the average temperature in Celsius, when the Celsius values are converted to Fahrenheit.")
```

#### 5. Find the standard deviation of the temperature lists (ÂºC and ÂºF). What is the relation between both standard deviations?

```python
temp_C_sqd = 0

for i in range(len(temperatures_C)):
    temp_C_sqd +=(temperatures_C[i] - avg_temp_C) ** 2
    temp_C_stdev = ((temp_C_sqd)/ len(temperatures_C)) ** 0.5
    
temp_F_sqd = 0

for i in range(len(temperatures_F)):
    temp_F_sqd += (temperatures_F[i] - avg_temp_F) ** 2
    temp_F_stdev = ((temp_F_sqd)/ len(temperatures_F)) ** 0.5
    
print ("Standard deviation of temperatures in Celsius is", round(temp_C_stdev,2),".")
print ("Standard deviation of temperatures in Fahrenheit is", round(temp_F_stdev,2),".")


if round(temp_F_stdev,2) == round(temp_C_stdev,2):
    print("They are exactly the same.")
elif round(temp_F_stdev,2) > round(temp_C_stdev, 2):
    print("The standard deviation in Fahrenheit is higher than the standard deviation in Celsius.")
elif round(temp_F_stdev,2) < round(temp_C_stdev, 2):
    print("standard deviation in Fahrenheit is lower than the standard deviation in Celsius.")   
```
